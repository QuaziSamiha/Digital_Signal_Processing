```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
from scipy.misc import electrocardiogram
from numpy.fft import fft
```

```python
eeg = electrocardiogram()
frequency = 300
time_data = np.arange(eeg.size) / frequency
plt.figure(figsize=(16,8))
plt.plot(time_data, eeg)
plt.xlabel("time in seconds")
plt.ylabel("EEG in ml Volts")
plt.show()
```

```python
pd.Series(eeg).plot(figsize=(16,8), lw=1, title='EEG')
plt.show()
```

```python
y = eeg[100000:101000]
x = np.arange(0,len(y),1)
plt.figure(figsize=(16,8))
plt.title("part of EEG signal",fontsize=10)
plt.plot(x,y)
plt.show()
```

```python
Y = fft(y)
N = len(Y)
n = np.arange(N)
T = N/256
freq = n/T
plt.figure(figsize=(16,8))
plt.title("fourier transform", fontsize=10)
plt.stem(freq, np.abs(Y))
plt.xlabel('Freq (Hz)', fontsize=8)
plt.ylabel('FFT Amplitude |X(freq)|', fontsize=8)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the length of the signal and the sampling frequency
signal_length = 5  # seconds
sampling_frequency = 1000  # Hz

# Generate a random EEG signal
time = np.linspace(0, signal_length, signal_length * sampling_frequency)
signal = np.random.normal(0, 1, len(time))

# Compute the Fourier transformation of the signal
fft_vals = np.fft.fft(signal)
freqs = np.fft.fftfreq(len(signal), 1/sampling_frequency)

# Take the absolute value of the complex Fourier coefficients
abs_fft = np.abs(fft_vals)

# Plot the power spectrum
plt.plot(freqs, abs_fft)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power')
plt.title('Power Spectrum of EEG Signal')
plt.show()

# Find the frequency with the highest power
max_freq_idx = np.argmax(abs_fft)
max_freq = freqs[max_freq_idx]

# Print the frequency with the highest power
print(f"The frequency with the highest power is {max_freq:.2f} Hz")
```
