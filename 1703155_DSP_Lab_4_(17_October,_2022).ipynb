```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# creating complex number
z = complex(4,3)
print(z)

# extracting complex and real part of the complex nubmer
print("Real part of " , z , " : " , np.real(z))
print("Complex part of ", z , " : ", np.imag(z))
```

```python
# conjugate number of a complex number
conj_z = np.conj(z)
print("conjugate number of c " , z ," complex number " , conj_z)
```

```python
# magnitude of complex number
magnitude = np.sqrt(np.real(z)**2 + np.imag(z)**2)
print("Magnitude : ", magnitude)
# phase of complex number
phase = np.angle(z)
print("Phase : " , phase)
```

```python
# ploting complex number
plt.figure(figsize = (8, 5))
plt.plot(np.real(z), np.imag(z), 'ro', markersize=10, label = "complex number")
plt.xlim([-8,8])
plt.ylim([-8,8])
plt.plot([-8,8],[0,0], 'b', lw=2)
plt.plot([0,0],[-8,8], 'b', lw=2)
plt.plot([0,np.real(z)], [0,np.imag(z)], 'g', lw='2')
plt.xlabel('Real Axis')
plt.ylabel("Imaginary Axis")
plt.title("Complex Plane")
plt.show()
```

```python
# Addition of complex number
z1 = complex(3,4)
z2 = complex(12,13)

add_z = complex(( np.real(z1)+ np.real(z2)), (np.imag(z1)+np.imag(z2)))
print(add_z)
```

```python
# complex sine wave
time = np.linspace(0, 10, 100)
frequency = 50;
Amplitude = 1;
theta = np.pi/2;
redu = -0.25

# complex signal generating
# complex_wave = Amplitude * np.exp(-1j * (2* np.pi * frequency * time + theta))
# reducible complex signal generating
complex_red_wave = Amplitude * np.exp(redu *time) * np.exp(-1j * (2* np.pi * frequency * time + theta))

plt.figure(figsize = (10, 5))
plt.plot(time, np.real(complex_red_wave), 'g-', linewidth = 2, label='real part')
plt.plot(time, np.imag(complex_red_wave), 'b-', linewidth = 2, label='imaginary part')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Complex sine wave')
plt.show()
```

```python
x1=np.array([1,2,3,1]) #origin at 0
x2=np.array([1,2,1,-1]) #origin at 1
x1_org=0
x2_org=1

new_org=x1_org+x2_org;

yn=np.convolve(x1,x2,mode="full")
plt.figure(figsize=(8,5))
plt.legend(fontsize=5)
print(yn , "Origin :",new_org)
x_axis=np.arange(-new_org,(x1.size+x2.size-new_org-1),1)
markerline,stemlines,baseline=plt.stem(x_axis,yn,label="Convulation Signal")
plt.setp(stemlines)
plt.show()
```
