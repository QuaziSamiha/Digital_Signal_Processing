```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# unit step signal
x = np.array([0, 0, 0, 1, 1, 1, 1])
n = np.array([-3, -2, -1, 0, 1, 2, 3])

# plotting unit step signal
plt.figure(figsize=(10,5))
plt.stem(n, x, label='Unit Step Signal')
plt.legend(fontsize=10)


# flipped and plotting unit step signal
flipped_x = x[::-1]
plt.figure(figsize=(10,5))
plt.stem(n, flipped_x, label='Flipped Unit Step Signal')
plt.legend(fontsize=10)

# shifting and plotting unit step signal
# x(n+2)
shifted_x = np.array([0,0,0,0,0,0,0])
shifted_n = np.array([-5,-4,-3,-2,-1,0,1])
for i in range (len(x)):
  if shifted_n[i] >= 0:
    shifted_x[i] = 1
  else:
    shifted_x[i] = 0

plt.figure(figsize=(10,5))
plt.stem(shifted_n, shifted_x, label='Shifted Unit Step Signal')
plt.legend(fontsize=10)

# scaling and plotting unit step signal
# x(2n)
for i in range (len(x)):
  print(i)


plt.show()
```
