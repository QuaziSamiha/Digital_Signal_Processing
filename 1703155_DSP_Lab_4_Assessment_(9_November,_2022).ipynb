````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
# z-transformation of a signal
# signal in time domain
n = np.linspace(0,10,100)
xn = 2 * np.sin(n) # x(n) = 2 sin(n)
plt.figure(figsize = (20,3))

plt.subplot(1,2,1)
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Time Domain (Continuous)')
plt.plot(n, xn, markersize=10)

plt.subplot(1,2,2)
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Time Domain (Discrete)')
plt.stem(n, xn, use_line_collection=True)

plt.show()
````

````python
# a particular rotating vector z^(-n) = A e^(-i*theta*n) for a particular z = A thete
A = 3
theta = np.pi/2

kernel = lambda A, theta: (A**(-n))*np.exp(-1j * (theta*n))
zn = kernel(A, theta)
for i in range(0, len(zn)):
  zn[i] = complex(zn[i])

plt.figure(figsize = (15, 5))
plt.subplot(1, 2, 1)
plt.xlabel('n')
plt.ylabel('x(n)')
plt.title('Time Domain')
plt.plot(n, np.real(zn), label="Real", markersize=15)
plt.plot(n, np.imag(zn), label='Imaginary', markersize=15)
plt.legend()

plt.subplot(1, 2, 2)
plt.title("Complex Domain")
plt.xlabel("Real")
plt.ylabel("Imaginary")
plt.scatter(np.real(x_nz_n), np.imag(x_nz_n), )
````
